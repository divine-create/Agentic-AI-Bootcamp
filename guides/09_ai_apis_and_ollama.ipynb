# AI APIs and Local Models

Overview of available AI APIs and local model options.

## Cloud APIs

### OpenAI
- GPT-4o, GPT-4o-mini
- API key required
- Cost: Pay per token

### Anthropic
- Claude 3.5, Claude 3
- API key required

### Google
- Gemini models
- API key required

## Local Models (Free)

### Ollama
Run models locally:
```bash
ollama serve
ollama pull llama3
```

### LM Studio
GUI for running local LLMs

## Setting API Keys

Create a `.env` file:
```
OPENAI_API_KEY=sk-...
ANTHROPIC_API_KEY=sk-ant-...
GOOGLE_API_KEY=...
```
