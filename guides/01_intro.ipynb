## What is Agentic AI?

Agentic AI refers to AI systems that can:
1. **Act autonomously** - Make decisions without constant human input
2. **Use tools** - Call external functions and APIs
3. **Plan and execute** - Break down complex tasks into steps
4. **Learn from feedback** - Iterate based on results

## LLM Provider: Gemini

In this course, we'll primarily use **Google Gemini** for practice. It's a powerful and cost-effective choice.

### Setting Up Your API Key

1. Go to [Google AI Studio](https://aistudio.google.com/app/apikey)
2. Create an API key
3. Add to your `.env` file:
```
GEMINI_API_KEY=your_key_here
```
4. Also set (some tools require both):
```
GOOGLE_API_KEY=your_key_here
```

````python
# Install required packages
!pip install google-genai python-dotenv
````

````python
import os
from dotenv import load_dotenv

# Load API key from .env
load_dotenv()
api_key = os.getenv("GEMINI_API_KEY")
print(f"API key loaded: {api_key is not None}")
````

````python
from google import genai

# Configure Gemini client
client = genai.Client(api_key=api_key)

# Test with a simple prompt
response = client.models.generate_content(
    model="gemini-2.0-flash",
    contents="What is Agentic AI in one sentence?"
)
print(response.text)
````
