## What is Agentic AI?

Agentic AI refers to AI systems that can:
1. **Act autonomously** - Make decisions without constant human input
2. **Use tools** - Call external functions and APIs
3. **Plan and execute** - Break down complex tasks into steps
4. **Learn from feedback** - Iterate based on results

### Traditional LLM vs Agent

| Traditional LLM | Agent |
|-----------------|-------|
| Responds to single prompt | Plans multi-step actions |
| No memory across sessions | Maintains state and context |
| Limited to training knowledge | Can use external tools |
| Passive | Proactive |

## LLM Provider: Google Gemini

In this course, we'll use **Google Gemini** as our primary LLM for practice.

### Why Gemini?

- **Cost-effective** - Generous free tier
- **Fast** - Excellent response times
- **Multimodal** - Text, images, video, audio
- **Context window** - Up to 1M tokens

### Setting Up Your API Key

1. Go to [Google AI Studio](https://aistudio.google.com/app/apikey)
2. Click "Create API Key"
3. Copy your key
4. Create a `.env` file in your project:
```
GEMINI_API_KEY=your_key_here
GOOGLE_API_KEY=your_key_here
```

````python
import os
from dotenv import load_dotenv

# Load API key from .env
load_dotenv()
api_key = os.getenv("GEMINI_API_KEY")

if api_key:
    print("✓ API key loaded successfully")
else:
    print("⚠ API key not found. Please set GEMINI_API_KEY in your .env file")
````

````python
from google import genai

# Configure Gemini client
client = genai.Client(api_key=api_key)

# Test with a simple prompt
response = client.models.generate_content(
    model="gemini-2.0-flash",
    contents="Explain what an AI agent is in one sentence."
)
print("Gemini Response:")
print(response.text)
````
