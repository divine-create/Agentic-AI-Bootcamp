```python
# Functions
def create_agent(name, role):
    return {"name": name, "role": role}

agent = create_agent("Assistant", "helper")
print(agent)
```

```python
# Classes
class Agent:
    def __init__(self, name, role):
        self.name = name
        self.role = role
    
    def act(self, task):
        return f"{self.name} doing {task}"

agent = Agent("Bot", "helper")
print(agent.act("work"))
```
