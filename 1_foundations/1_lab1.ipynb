```python
# The google-genai SDK is the official Python client for Gemini
# If you get an ImportError here, run: uv add google-genai

from google import genai

# Create the client — it reads GEMINI_API_KEY from your environment automatically
client = genai.Client()
```

```python
# Your very first Gemini call!
# gemini-2.0-flash is fast, cheap, and very capable

response = client.models.generate_content(
    model="gemini-2.0-flash",
    contents="What is 2 + 2?"
)

print(response.text)
```

## Congratulations — you just called an LLM!

Now let's make it do something more interesting. We'll build a **3-step AI pipeline**:

1. Ask Gemini to generate a challenging IQ-style question
2. Ask Gemini to answer that question
3. Display the result with nice formatting

```python
# Step 1: Generate a hard question

step1 = client.models.generate_content(
    model="gemini-2.0-flash",
    contents="Propose a hard, challenging question to assess someone's IQ. Respond only with the question itself."
)

question = step1.text
print("Generated question:", question)
```

```python
# Step 2: Answer the question
# Note: we pass the question as the contents of the second call

step2 = client.models.generate_content(
    model="gemini-2.0-flash",
    contents=question
)

answer = step2.text
print(answer)
```

```python
# Step 3: Display nicely with Markdown rendering

from IPython.display import Markdown, display

display(Markdown(answer))
```

## Exercise

Build this 3-step business idea pipeline:

1. **Step 1** — Ask Gemini to pick a business area that might benefit from Agentic AI
2. **Step 2** — Ask Gemini to describe a specific pain-point in that industry
3. **Step 3** — Ask Gemini to propose an Agentic AI solution to that pain-point

Each call should use the output of the previous one as input. Save your answers to `exercises/your_name/` once done!

```python
# Step 1: Pick a business area
step1 = client.models.generate_content(
    model="gemini-2.0-flash",
    contents="Pick one business industry that could most benefit from Agentic AI. Respond with only the industry name and a one-sentence reason."
)
business_area = step1.text
print(business_area)
```

```python
# Step 2: Identify a pain point — use business_area in the prompt
step2 = client.models.generate_content(
    model="gemini-2.0-flash",
    contents=f"Given this business area: {business_area}\n\nDescribe one significant pain-point that is currently difficult to solve."
)
pain_point = step2.text
print(pain_point)
```

```python
# Step 3: Propose an Agentic AI solution
step3 = client.models.generate_content(
    model="gemini-2.0-flash",
    contents=f"Business area: {business_area}\n\nPain point: {pain_point}\n\nPropose a concrete Agentic AI solution with specific agents, tools and workflow."
)
display(Markdown(step3.text))
```
